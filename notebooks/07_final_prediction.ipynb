백테스트에서 **최고 수익률 모델**로 실제 미래 가격을 예측합니다.

- 향후 30~60일의 일별 예상 고가 / 저가
- 스윙매매 진입/청산 가격 테이블
- 예상 수익 시뮬레이션

```python
!pip install xgboost pandas-ta yfinance scikit-learn tensorflow -q
!apt-get install -y fonts-nanum -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas_ta as ta
import yfinance as yf
import joblib
import warnings
import os
from datetime import datetime
import tensorflow as tf
from sklearn.preprocessing import RobustScaler

warnings.filterwarnings('ignore')

fm.fontManager.addfont('/usr/share/fonts/truetype/nanum/NanumGothic.ttf')
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (14, 6)

INITIAL_CAPITAL = 100_000
FORECAST_DAYS = 30

try:
    summary = pd.read_csv('/content/results/backtest_summary.csv')
    BEST_MODEL = summary.sort_values('total_return(%)', ascending=False).iloc[0]['model']
    print(f'최고 성능 모델: {BEST_MODEL}')
except:
    BEST_MODEL = 'XGBoost'
    print(f'기본 모델 사용: {BEST_MODEL}')

print('한글 폰트: NanumGothic')
```

```python
tsll_latest = yf.download('TSLL', start='2022-08-09', progress=False)
tsla_latest = yf.download('TSLA', start='2022-08-09', progress=False)
tsll_latest.columns = [col[0] if isinstance(col, tuple) else col for col in tsll_latest.columns]
tsla_latest.columns = [col[0] if isinstance(col, tuple) else col for col in tsla_latest.columns]
print(f'TSLL 최신 데이터: {tsll_latest.index[-1].date()} 까지 ({len(tsll_latest)}일)')
```

```python
def add_features(df):
    df = df.copy()
    df['sma_20']  = ta.sma(df['Close'], length=20)
    df['sma_50']  = ta.sma(df['Close'], length=50)
    df['ema_12']  = ta.ema(df['Close'], length=12)
    df['ema_26']  = ta.ema(df['Close'], length=26)
    df['price_vs_sma20'] = (df['Close'] - df['sma_20']) / df['sma_20']
    df['price_vs_sma50'] = (df['Close'] - df['sma_50']) / df['sma_50']
    df['rsi_14']  = ta.rsi(df['Close'], length=14)
    df['rsi_7']   = ta.rsi(df['Close'], length=7)
    macd = ta.macd(df['Close'])
    if macd is not None:
        df['macd']        = macd.iloc[:,0]
        df['macd_signal'] = macd.iloc[:,1]
        df['macd_hist']   = macd.iloc[:,2]
    bb = ta.bbands(df['Close'], length=20)
    if bb is not None:
        df['bb_upper'] = bb.iloc[:,0]
        df['bb_mid']   = bb.iloc[:,1]
        df['bb_lower'] = bb.iloc[:,2]
        df['bb_width'] = (df['bb_upper'] - df['bb_lower']) / df['bb_mid']
        df['bb_pct']   = (df['Close'] - df['bb_lower']) / (df['bb_upper'] - df['bb_lower'])
    df['atr_14']  = ta.atr(df['High'], df['Low'], df['Close'], length=14)
    df['atr_7']   = ta.atr(df['High'], df['Low'], df['Close'], length=7)
    df['atr_pct'] = df['atr_14'] / df['Close']
    df['obv']     = ta.obv(df['Close'], df['Volume'])
    df['obv_ema'] = ta.ema(df['obv'], length=20)
    df['volume_ratio'] = df['Volume'] / df['Volume'].rolling(20).mean()
    df['return_1d']  = df['Close'].pct_change(1)
    df['return_3d']  = df['Close'].pct_change(3)
    df['return_5d']  = df['Close'].pct_change(5)
    df['return_10d'] = df['Close'].pct_change(10)
    for lag in [1, 2, 3, 5]:
        df[f'close_lag{lag}']  = df['Close'].shift(lag)
        df[f'return_lag{lag}'] = df['return_1d'].shift(lag)
        df[f'high_lag{lag}']   = df['High'].shift(lag)
        df[f'low_lag{lag}']    = df['Low'].shift(lag)
    df['day_of_week']  = df.index.dayofweek
    df['month']        = df.index.month
    df['week_of_year'] = df.index.isocalendar().week.astype(int)
    df['hl_ratio']       = df['High'] / df['Low']
    df['co_ratio']       = df['Close'] / df['Open']
    df['intraday_range'] = (df['High'] - df['Low']) / df['Open']
    return df

tsll_feat = add_features(tsll_latest)
tsla_feat = add_features(tsla_latest)
tsla_cols = ['Close', 'return_1d', 'rsi_14', 'atr_pct', 'macd_hist', 'volume_ratio']
tsla_sub  = tsla_feat[[c for c in tsla_cols if c in tsla_feat.columns]].copy()
tsla_sub.columns = [f'tsla_{c}' for c in tsla_sub.columns]
combined = tsll_feat.join(tsla_sub, how='left').dropna()
TARGET_COLS  = ['target_high', 'target_low', 'target_close', 'target_return']
FEATURE_COLS = [c for c in combined.columns if c not in TARGET_COLS]
print(f'최신 특성 수: {len(FEATURE_COLS)}')
```

```python
xgb_high = joblib.load('/content/models/xgb_high.pkl')
xgb_low  = joblib.load('/content/models/xgb_low.pkl')

forecast_high_list = []
forecast_low_list  = []
forecast_dates     = []
current_df = combined.copy()

for day in range(FORECAST_DAYS):
    last_row = current_df[FEATURE_COLS].fillna(0).iloc[[-1]]
    pred_h = xgb_high.predict(last_row)[0]
    pred_l = xgb_low.predict(last_row)[0]
    pred_c = (pred_h + pred_l) / 2
    last_date = current_df.index[-1]
    next_date = last_date + pd.offsets.BDay(1)
    forecast_high_list.append(pred_h)
    forecast_low_list.append(pred_l)
    forecast_dates.append(next_date)
    new_row = current_df.iloc[-1:].copy()
    new_row.index = [next_date]
    new_row['Close'] = pred_c
    new_row['High']  = pred_h
    new_row['Low']   = pred_l
    new_row['Open']  = current_df['Close'].iloc[-1]
    current_df = pd.concat([current_df, new_row])
    current_df = add_features(current_df)

forecast_df = pd.DataFrame({
    'date':      forecast_dates,
    'pred_high': forecast_high_list,
    'pred_low':  forecast_low_list,
    'pred_mid':  [(h+l)/2 for h, l in zip(forecast_high_list, forecast_low_list)]
})
print('예측 완료!')
print(forecast_df.head(10).to_string(index=False))
```

```python
last_90 = combined.tail(90)
fig, ax = plt.subplots(figsize=(16, 7))
ax.plot(last_90.index, last_90['Close'], color='white', lw=2, label='실제 종가 (과거)')
ax.fill_between(last_90.index, last_90['Low'], last_90['High'], alpha=0.15, color='white')
ax.plot(forecast_df['date'], forecast_df['pred_mid'], color='#ffd700', lw=2, ls='--', label='예측 중간값 (미래)')
ax.fill_between(forecast_df['date'], forecast_df['pred_low'], forecast_df['pred_high'],
                alpha=0.25, color='#ffd700', label='예측 고가/저가 범위')
ax.axvline(x=combined.index[-1], color='#ef5350', lw=2, ls=':', label='예측 시작점 (현재)')
ax.set_title(f'TSLL 미래 {FORECAST_DAYS}거래일 예측 — {BEST_MODEL} 모델', fontsize=14)
ax.set_ylabel('가격 (USD)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('/content/final_prediction.png', dpi=150, bbox_inches='tight')
plt.show()

swing_table = forecast_df.copy()
swing_table['매수주문가'] = (swing_table['pred_low']  * 0.995).round(2)
swing_table['매도목표가'] = (swing_table['pred_high'] * 1.005).round(2)
swing_table['예상수익률'] = ((swing_table['매도목표가'] - swing_table['매수주문가']) / swing_table['매수주문가'] * 100).round(2)
swing_table['date'] = swing_table['date'].dt.strftime('%Y-%m-%d')
print(swing_table[['date', '매수주문가', '매도목표가', '예상수익률']].to_string(index=False))
```

```python
avg_daily_return = swing_table['예상수익률'].mean()
annual_projected = (1 + avg_daily_return / 100) ** 252 - 1
print('\n' + '='*50)
print('예상 수익 요약')
print('='*50)
print(f'시드머니:              ${INITIAL_CAPITAL:,}')
print(f'예측 모델:             {BEST_MODEL}')
print(f'예측 기간:             {FORECAST_DAYS}거래일')
print(f'평균 일별 예상 수익률: {avg_daily_return:.2f}%')
print(f'복리 연환산 수익률:    {annual_projected*100:.1f}%')
print(f'1년 후 예상 자산:      ${INITIAL_CAPITAL * (1 + annual_projected):,.0f}')
print('='*50)
print('주의: 이 예측은 과거 패턴 기반이며 실제 투자 결과를 보장하지 않습니다.')
os.makedirs('/content/results', exist_ok=True)
forecast_df.to_csv('/content/results/final_forecast.csv', index=False)
swing_table.to_csv('/content/results/swing_trade_plan.csv', index=False)
print('최종 예측 저장 완료!')
```
